```python
b.plot(y="Generosity",x='Healthy life expectancy', figsize=(20,3), kind = 'scatter')
```

```python
b.plot(y="Social support",x='Healthy life expectancy', figsize=(20,3), kind = 'scatter')
```

Here we can see a more obvious correlation between Social support and life expectancy, the more a society tends to support eachother the higher the life expectancy is.


```python
b.plot(y="Logged GDP per capita",x='Healthy life expectancy', figsize=(20,3), kind = 'scatter')
```
