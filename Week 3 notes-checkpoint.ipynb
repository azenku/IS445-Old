```python
#step 2 is making a function (WIDGETIZING)
def plot_colormap(col_map_name):
    
    col_map = plt.get_cmap(col_map_name) #hard coded colormap

    fig = plt.figure(figsize= (4,3)) # i will make a figure axis but not specify axes
    #                left, bottom, width, height
    ax = fig.add_axes([0.0, 0.15,    1.0,   0.8])
    
    colors =col_map(color_steps)[:,:3]
    ax.plot(color_steps, colors[:,0], '-r')
    ax.plot(color_steps, colors[:,1], '-g')
    ax.plot(color_steps, colors[:,2], '-b')
    ax.set_ylabel("Relative Strength of RGB")

    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)

    #                           left, bottom, width, height
    ax_colorbar = fig.add_axes([0.0,    0.0,   1.0,  0.2])
    matplotlib.colorbar.ColorbarBase(ax_colorbar, cmap = col_map, orientation = 'horizontal')
    ax_colorbar.xaxis.set_visible(False)
    plt.show()
```

```python
plot_colormap("Blues")
```

```python
#FINAL STEP TO (WDIGITIZING)
@ipywidgets.interact(col_map_name = plt.colormaps())
def plot_colormap(col_map_name):
    
    col_map = plt.get_cmap(col_map_name) #hard coded colormap

    fig = plt.figure(figsize= (4,3)) # i will make a figure axis but not specify axes
    #                left, bottom, width, height
    ax = fig.add_axes([0.0, 0.15,    1.0,   0.8])
    
    colors =col_map(color_steps)[:,:3]
    ax.plot(color_steps, colors[:,0], '-r')
    ax.plot(color_steps, colors[:,1], '-g')
    ax.plot(color_steps, colors[:,2], '-b')
    ax.set_ylabel("Relative Strength of RGB")

    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)

    #                           left, bottom, width, height
    ax_colorbar = fig.add_axes([0.0,    0.0,   1.0,  0.2])
    matplotlib.colorbar.ColorbarBase(ax_colorbar, cmap = col_map, orientation = 'horizontal')
    ax_colorbar.xaxis.set_visible(False)
    plt.show()
```

```python
@ipywidgets.interact(col_map_name = plt.colormaps(), sym_log=True)
def plot_michigan_color(col_map_name,sym_log):
    if sym_log:
        plt.imshow(michigan, cmap=col_map_name, norm = matplotlib.colors.SymLogNorm(10)) #hardcoded map, lets widgetize
    else:
        plt.imshow(michigan, cmap=col_map_name) #hardcoded map, lets widgetize
    plt.clim(-352,352)
    plt.colorbar()
    plt.show()
```
